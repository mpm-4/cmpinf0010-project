## Project Introduction and Approach

This project aims to analyze neighborhood farmers market data to identify which neighborhood offers the best access to fresh and local food. Using the available data, I focused on quantitative metrics such as the number of farmers markets in each neighborhood (city) and the geographic spread of the markets within that neighborhood.

## Metric Description and Dataset

The metric developed to evaluate the best neighborhood combines two key features from the farmers market dataset:

- **Number of Farmers Markets:** The count of farmers markets within each neighborhood, serving as an indicator of access and availability to local food sources.

- **Geographic Spread:** The range of latitude and longitude values for the markets in each neighborhood, reflecting how dispersed the markets are and potentially the accessibility for residents throughout the neighborhood.

These features balance quantity and coverage, as having many markets concentrated in one area may differ from having fewer markets spread across the neighborhood.

## Dataset

The dataset used includes detailed information about farmers markets in Allegheny County, such as:

- Market name  
- Market type (Farm Market or Farm Stand)  
- Neighborhood (city)  
- Address, latitude, and longitude  
- County  

The combined metric ranks neighborhoods by both the number of markets and their geographic spread, producing an overall neighborhood ranking that reflects both the availability and accessibility of farmers markets.



```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('farmersmarket.csv')

# Column to group byneighborhood
neighborhood_col = 'city'

# Count number of farmers markets in each neighborhood
market_counts = df[neighborhood_col].value_counts().reset_index()
market_counts.columns = ['Neighborhood', 'NumMarkets']

# Sort neighborhoods by amt of markets
market_counts = market_counts.sort_values('NumMarkets', ascending=False)


print(market_counts)

# Display best neighborhood
best = market_counts.iloc[0]
print(f"Best neighborhood by number of farmers markets: {best['Neighborhood']} with {best['NumMarkets']} markets")

# Visualization: Bar chart of top 10 neighborhoods 
top10 = market_counts.head(10)
plt.figure(figsize=(12,6))
plt.bar(top10['Neighborhood'], top10['NumMarkets'], color='green')
plt.title('Top 10 Neighborhoods by Number of Farmers Markets')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Farmers Markets')
plt.xticks(rotation=45)
plt.show()
```

### Conclusion

Based on our analysis of farmers markets in Allegheny County, **Pittsburgh emerges as the best neighborhood for access to fresh and local food**. This conclusion is based on the number of farmers markets in each neighborhood, with Pittsburgh having the highest number of markets, providing residents with multiple options for obtaining fresh, local produce.

This data-driven determination of the best neighborhood focuses on market availability. While we did not analyze exact distribution patterns within each neighborhood, the high number of markets suggests that residents have more opportunities to access local food compared to other neighborhoods. Personal preferences may vary, but our metric provides an objective measure of access based on quantity.
